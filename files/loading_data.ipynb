## Price filtering and cleaning

```python
# Drop 100% missing and very high missing columns
cols_to_drop = [
    'calendar_updated',      # 100% missing
    'license',               # 85% missing, not useful for analysis
    'neighborhood_overview', # 47% missing, free text not needed
    'estimated_revenue_l365d' # 40% missing, we'll calculate our own
]

df = df.drop(columns=cols_to_drop)
print(f"Shape after dropping useless columns: {df.shape}")
```

```python
# Remove listings with no price — can't analyze them
df = df.dropna(subset=['price'])
print(f"Shape after dropping null prices: {df.shape}")
print(f"Price sample:\n{df['price'].head(10)}")
```

```python
df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)
```

```python
# Show how many listings are above $10,000
total = len(df)
above_10k = len(df[df['price'] > 10000])
pct = (above_10k / total) * 100

print(f"Listings above $10,000: {above_10k} ({pct:.2f}% of total)")
print(f"\nSample of extreme prices:")
print(df[df['price'] > 10000][['name', 'neighbourhood_cleansed', 'room_type', 'price']].head(10))
```

Listings priced above $10,000 per night are removed for the following reasons:

1. Statistical outliers — NYC median Airbnb price is around $150-200/night. 
   Prices above $10,000 are extreme outliers that would heavily skew 
   averages, visualizations, and any regression models.

2. Likely data errors — Some hosts enter incorrect prices (e.g. $10,000 
   instead of $100) or use placeholder values. These are not representative 
   of real market pricing.

3. Different market segment — Ultra-luxury rentals above $10,000/night 
   represent a completely different market (private mansions, penthouses) 
   that is not relevant to a typical Airbnb market analysis.

4. Volume is negligible — Listings above $10,000 represent less than 0.1% 
   of all listings, so removing them has no meaningful impact on the dataset.

```python
df = df[(df['price'] > 0) & (df['price'] < 10000)]

print(f"Price range: ${df['price'].min()} — ${df['price'].max()}")
print(f"Median price: ${df['price'].median()}")
print(f"Rows remaining: {len(df):,}")
```

## Fix date columns 

```python
date_cols = ['host_since', 'first_review', 'last_review']
for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')

print("Date columns fixed:")
print(df[date_cols].dtypes)
```

```python
df.head()
```

## Fixing percentage Rates

```python
for col in ['host_response_rate', 'host_acceptance_rate']:
    df[col] = df[col].replace('%', '', regex=True)
    df[col] = pd.to_numeric(df[col], errors='coerce')

print(df[['host_response_rate', 'host_acceptance_rate']].describe())
```

## Fill numeric nulls

```python
# Reviews — 0 for listings with no reviews yet
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)

# Beds and bedrooms — fill with median
df['beds'] = df['beds'].fillna(df['beds'].median())
df['bedrooms'] = df['bedrooms'].fillna(df['bedrooms'].median())
df['bathrooms'] = df['bathrooms'].fillna(df['bathrooms'].median())

print("Numeric nulls filled")
print(df[['beds', 'bedrooms', 'bathrooms']].isnull().sum())
```

## Categorical nulls

```python
df['neighbourhood'] = df['neighbourhood'].fillna(df['neighbourhood_cleansed'])
df['host_about'] = df['host_about'].fillna('Not provided')
df['description'] = df['description'].fillna('Not provided')
df['bathrooms_text'] = df['bathrooms_text'].fillna('Not provided')

print("Categorical nulls filled")
```

## Fixing is boolean columns

```python
# Convert t/f to True/False boolean
for col in ['host_is_superhost','instant_bookable','has_availability','host_has_profile_pic','host_verifications','host_identity_verified']:
    df[col] = df[col].map({'t': True, 'f': False})
    df[col] = df[col].fillna(False)

# print(df['host_is_superhost'].value_counts())
```

## Remaining nulls

```python
remaining_nulls = df.isnull().sum()
remaining_nulls = remaining_nulls[remaining_nulls > 0]

print(f"Final shape: {df.shape}")
print(f"\nRemaining nulls:\n{remaining_nulls}")
```

```python
# calendar_updated and license are not useful for analysis
df = df.drop(columns=['calendar_updated', 'license'], errors='ignore')
print("Dropped calendar_updated and license")
```

```python
# Host location and neighbourhood are optional profile fields — not critical
df['host_location'] = df['host_location'].fillna('Not provided')
df['host_neighbourhood'] = df['host_neighbourhood'].fillna('Not provided')
df['neighborhood_overview'] = df['neighborhood_overview'].fillna('Not provided')

# Host response fields — unknown means they haven't set it
df['host_response_time'] = df['host_response_time'].fillna('Unknown')
df['host_response_rate'] = df['host_response_rate'].fillna(0)
df['host_acceptance_rate'] = df['host_acceptance_rate'].fillna(0)

print("Host nulls filled")
```

```python
# Listings with no reviews will have null review scores — this is valid
# Keep them but flag them separately
review_cols = [
    'review_scores_rating', 'review_scores_accuracy',
    'review_scores_cleanliness', 'review_scores_checkin',
    'review_scores_communication', 'review_scores_location',
    'review_scores_value'
]

# Fill with 0 to indicate no reviews yet
for col in review_cols:
    df[col] = df[col].fillna(0)

# Also fill first/last review dates
df['first_review'] = df['first_review'].fillna(pd.NaT)
df['last_review'] = df['last_review'].fillna(pd.NaT)

print("Review nulls filled")
```

```python
# These columns have very few nulls (<5 rows) — safe to drop
tiny_null_cols = [
    'host_name', 'host_since', 'host_thumbnail_url',
    'host_picture_url', 'host_listings_count', 
    'host_total_listings_count'
]
df = df.dropna(subset=tiny_null_cols)
print(f"Shape after dropping tiny null rows: {df.shape}")
```

```python
remaining = df.isnull().sum()
remaining = remaining[remaining > 0]

if len(remaining) == 0:
    print("✅ No nulls remaining — data is clean!")
else:
    print(f"Remaining nulls:\n{remaining}")

print(f"\nFinal clean dataset shape: {df.shape}")
print(f"Total listings: {len(df):,}")
```
